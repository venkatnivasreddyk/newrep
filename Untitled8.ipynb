```python
from sklearn.datasets import load_iris
import numpy as np
from sklearn.ensemble import RandomForestClassifier
np.random.seed(0)
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
data=sns.load_dataset('iris')
```

```python
iris=load_iris()
```

```python
data.head(100)
```

```python
lb=LabelEncoder()
data['is_train']=lb.fit_transform(data['species'])
```

```python
data
```

```python
data['train']=np.random.uniform(0,1,len(data))<=.80
data.head()
```

```python
train=data[data['train']==True]
test=data[data['train']==False]
print('number of observations in the training data:',len(train))
print('number of observations in the test data:',len(test))
```

```python
data
```

```python
feautures=data.columns[:4]
feautures
```

```python
y=pd.factorize(train['species'])[0]
```

```python
y
```

```python
clf=RandomForestClassifier(n_jobs=2,random_state=0)
clf.fit(train[feautures],y)
```

```python
clf.predict(test[feautures])
```

```python
test[feautures]
```

```python
clf.predict_proba(test[feautures])[10:20]
```

```python
preds=iris.target_names[clf.predict(test[feautures])]
preds[0:5]
```

```python
pd.crosstab(test['species'],preds,rownames=['Actual Species'],colnames=['Predicted Species'])
```
